```python
import numpy as np
from matplotlib import pyplot as plt
import random as rd

inputLayerSize, hiddenLayerSize, outputLayerSize = 2, 2, 2

wh = np.random.uniform(size=(inputLayerSize, hiddenLayerSize))
wo = np.random.uniform(size=(hiddenLayerSize,outputLayerSize))

print(wh)
print()
print(wo)
print()
#fig = plt.figure(figsize=(9,7))
#ax = Axes3D(fig)

def sigmoid (x): 
    return 1/(1 + np.exp(-x)) 

data1=[]
for i in range (0,100):
    i=rd.randint(0,1)
    data1.append(i)
data2=[]
for i in range (0,100):
    i=rd.randint(0,1)
    data2.append(i) 

data = np.concatenate((data1, data2))
data=np.reshape(data,[100,2])
#print(data)
#ax.scatter(data1[:,0], data1[:,1], data1[:,2],c='red', marker = '*')
#ax.scatter(data2[:,0], data2[:,1], data2[:,2],c='g', marker = 'o')
#plt.show()

y=[0,1]

#for i in range(epochs):
e=[]
err=[]
h=np.dot(data,wh)
H = sigmoid(h)
yh=np.dot(H, wo)
yhat = sigmoid(yh)   

E=y-yhat

#print(yhat)

for i in range(0,100):
    err.append(0.5*(y-yhat[i])**2)
    #e.append(0.5*np.sum(err))

learn=0.2
#for i in range (0,100):
    #print(err[i])
#weight uodate for output layer to hidden layer

sum=[]
epoch=[]
for i in range (0,100):
    epoch.append(i)
    ax=err[i][0]+err[i][1]
    sum.append(ax)

plt.xlabel('Iteration') 
plt.ylabel('Sum of squared errors') 
plt.plot(epoch,sum)
plt.show()
```

```python
epoch=[]
e=[]
for k in range(0,1000):
    for i in range(0,100):
        de=-(y-yhat[i][0])
        dy=yhat[i][0]*(1-yhat[i][0])
        dw=yhat[i][0]

        dw0=de*dy*dw

        wo1=wo-learn*dw0

    
    #weight uodate for hidden layer to input layer
    for i in range(0,100):
        di=yh[i]*wo1*(1-yhat[i])*data[i]
        do=wh-learn*di

#print(wo1)
#print()        
#print(do)
    
    h=np.dot(data,do)
    H = sigmoid(h)
    yh=np.dot(H, wo1)
    yhat = sigmoid(yh)   

    E=y-yhat

    #print(yhat)

    for i in range(0,100):
        err[i]=0.5*(y-yhat[i])**2
        #e.append(0.5*np.sum(err))
    sum=0
    
    for i in range (0,100):
        ax=err[i][0]+err[i][1]
        sum+=ax
    e.append(sum)
    #print(sum)
    epoch.append(k)
    
    wo=wo1
    wh=do
    
    
plt.xlabel('Iteration') 
plt.ylabel('Sum of squared errors') 
plt.plot(epoch,e)
plt.show()
```
